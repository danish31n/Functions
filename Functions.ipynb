## Q1 What is the difference between a function and a method in Python?
- In Python, both functions and methods are callable objects, but they differ in their context and how they are used:

####1. Function:
- A function is a block of reusable code that performs a specific task.
- Functions are defined using the def keyword and can be called independently.
- They can exist outside of a class and operate on the arguments passed to them.

      #Example of a function:
      def greet(num):
           return f"Hello, {num}!"
           print(greet("Danish"))  # Output: Hello, Danish!

####2. Method:
- A method is a function that is associated with an object or class.
- Methods are defined within a class and are called on an instance (or the class itself, in the case of class methods).
- Methods implicitly take the instance (self) or class (cls) as their first parameter, allowing them to operate on instance or class-specific data.

           #Example of function:
           class Person:
               def __init__(self, num):    # Instance variable
               self.num = num

          def greet(self):                       # Instance method
               return f"Hello, {self.num}!"
           
           # Creating an instance of the Person class
           p = Person("Danish")

           # Calling the greet method on the instance
           print(p.greet())  # Output: Hello, Danish!

All methods are functions, but not all functions are methods.


















##Q2 Explain the concept of function arguments and parameters in Python.

- In Python, function arguments and parameters are related concepts that define how functions accept input values.


####  Function Parameters:-
- Parameters are the variables defined in a function's definition. They act as placeholders for the values that the function will receive when it is called.
When you define a function, you specify the parameters that it will accept.

- Example of function parameters:
      def greet(num):  # 'name' is a parameter
          return f"Hello, {num}!"

- In this case, num is the parameter of the `greet()` function. It's a placeholder that will be replaced by an actual value when the function is called.


#### Function Arguments:-
- Arguments are the actual values you pass to a function when you call it. These values are assigned to the corresponding parameters of the function.

- Example of function arguments:

        g = greet("Danish")  # "Danish" is the argument passed to the 'num' parameter
        print(g)  # Output: Hello, Danish!

##Q4 What is the purpose of the `return` statement in a Python function?

- The `return` statement is to send a result back from a function to the caller. When a function is called, it executes its code, and the `return` statement allows it to send a value (or multiple values) back to where the function was invoked.

- The `return` statement immediately ends the execution of the function. After a `return` is encountered, the function execution halts, and no further code in the function is executed. If `return` is called without any value, the function will `return` None.

- You can `return` multiple values in Python by separating them with commas. These values are returned as a tuple, which can be unpacked when received.


##Q5 What are iterators in Python and how do they differ from iterables?

#### Iterators:

   - An iterator is an object in Python that represents a stream of data. It allows you to traverse through its elements one by one. It implements two essential methods:

- `iter()` - Returns the iterator object itself.
- `next()` - Returns the next element in the sequence and raises a StopIteration exception when there are no more elements.


####Key Differences Between Iterables and Iterators

- An iterable is an object that can be iterated over (e.g., a list, tuple, string). It provides an iterator when passed to the `iter()` function.

- While an iterator is the object that performs the actual iteration and tracks the state of the iteration (e.g., it produces the next element when `next()` is called).

#####Methods:

- Iterable must implement the `iter()` method.
- Iterator must implement both the `iter()` and `next()` methods.

#####Use:

- You can loop over an iterable using a for loop, which automatically uses the iterator internally.
- You manually use an iterator to retrieve elements using `next()` until it raises 'StopIteration'.





##Q6 Explain the concept of generators in Python and how they are defined.

- A generator is a special type of iterator in Python that allows you to iterate over a sequence of values lazily (i.e., on demand), one at a time, without storing the entire sequence in memory. This is particularly useful for working with large data sets or streams of data, as it saves memory by producing items only when required.

####Generators can be defined as:

- A generator function is a regular Python function, but instead of returning a value using return, it uses the yield keyword. When the generator function is called, it returns a generator object, not the actual value. The yield statement produces a value and pauses the functionâ€™s execution, saving its state. The function can be resumed later from where it left off.

   - Example
            def sq(n):
                for i in range(n):
                    yield i**2  # Yield each squared value one by one
            
            # Convert the generator to a list and print the result
            print(list(sq(10)))
             

- The yield statement is now inside the for loop, which allows the function to yield each squared number one at a time.

- list(sq(10)) will iterate through the generator and collect the values into a list.









##Q7 What are the advantages of using generators over regular functions?

- Memory Efficiency: Generators don't store all values at once, making them ideal for large or infinite data sequences.
- Performance: Faster for large data since no intermediate data structures need to be built
- Infinite Sequences: You can work with infinite data streams without running out of memory.
- Cleaner and Simpler Code: The yield keyword helps you avoid managing extra state, making code more concise.


####Why Not Use Regular Functions?
- Higher Memory Usage: Regular functions that return a full list or collection require storing all values in memory, which can lead to performance bottlenecks when dealing with large datasets.
- Slower Performance: Since regular functions compute everything upfront, they are less efficient for large-scale data processing compared to the lazy evaluation provided by generators.
- Inability to Handle Infinite Sequences: Regular functions cannot handle infinite data streams because they attempt to generate and store all data at once, leading to impractical memory consumption.








##Q8 What is a lambda function in Python and when is it typically used?
- A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions, which are defined using the `def` keyword, lambda functions are often used for short-term tasks and do not require a name.

#### Syntax:

           lambda arguments: expression


Example:
        # A simple lambda function that adds two numbers
        add = lambda x, y: x + y
        print(add(3, 5))  # Output: 8
        

####Common Use Cases:
- In-line Function Definitions: Lambda functions are often used when a simple function is required temporarily, such as in a `map()`, `filter()`, or `sorted()` function, where you don't want to define a full function with def.

Example with `map()`:

        num = [1, 2, 3, 4, 5]
        sq = list(map(lambda x: x**2, num))  # Square each number
        print(sq)  # Output: [1, 4, 9, 16, 25]
              

      








##Q9 Explain the purpose and usage of the `map()` function in Python.


- The `map()` function in Python is a built-in function that allows you to apply a specified function to each item in an iterable (like a list, tuple, etc.) and return an iterator (a map object) that yields the results. It is a convenient way to transform elements in an iterable without the need for an explicit loop.

Syntax:
      
       map(function, iterable)
Example:

       numbers = [1, 2, 3, 4, 5]
       sq = map(lambda x: x**2, numbers)  # Square each number in the list
       
       # Convert the map object to a list
       sq_list = list(sq)
       print(sq_list)  # Output: [1, 4, 9, 16, 25]
       
####Usage of the `map()` function:

- `map()` is typically used to transform or modify the elements of an iterable based on a function, like squaring numbers, converting strings to uppercase, etc.
- `map()` follows a functional programming style, where you pass a function to process data instead of using loops.
- `map()` can be more efficient than using a loop, especially when combined with lambda functions, as it processes elements lazily (it returns an iterator).
- It can make code more concise and expressive when transforming data, but it can also reduce readability if overused or made too complex.







##Q10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- In Python, `map()`, `reduce()`, and `filter()` are built-in functions that are used for functional programming. They each serve different purposes when working with iterables (like lists or tuples).

####1. `map()`:
- Purpose: The `map()` function applies a given function to all items in an iterable (e.g., list, tuple) and returns an iterator that produces the results.
- Syntax:
      map(function, iterable)
- How it works: The function is applied to each element in the iterable (or iterables) and returns a new iterable with the results.


####2. `filter()`:
- Purpose: The `filter()` function filters elements from an iterable based on a given condition (a function that returns a boolean value). It returns an iterator containing only the elements that satisfy the condition.
- Syntax:
          filter(function, iterable)
- How it works: The function is applied to each element in the iterable. If the result of the function is True, that element is included in the returned iterator; otherwise, it is excluded.

####3. `reduce()` (from the functools module):
- Purpose: The `reduce()` function applies a rolling computation (a binary function) to the elements of an iterable, reducing the iterable to a single cumulative value.
- Syntax:
          reduce(function, iterable[, initial])
- How it works: The function is applied cumulatively to the elements of the iterable, from left to right. The result of the first application is passed to the next application, and so on, until a single result is produced.

```python
#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


def fib(n):
  """
  This function generates the Fibonacci sequence up to a specified number of terms.
  """
  # Initialize the first two Fibonacci numbers, 0 and 1
  a=0
  b=1
  for i in range(n):     # Loop 'n' times to generate 'n' Fibonacci numbers
    yield a              # Yield the current Fibonacci number 'a'
    a,b=b,a+b             # Update the values of 'a' and 'b' to the next two Fibonacci numbers

# Create the Fibonacci sequence generator for the first 10 terms
x=fib(10)          # This will generate a Fibonacci sequence up to 10 terms


# Using next() to get and print the Fibonacci numbers one by one
print(next(x))      # Output: 0
print(next(x))    # Output: 1
print(next(x))   # Output: 1
print(next(x))    # Output: 2
print(next(x))     # Output: 3
# and so on...

# The process continues and prints the next Fibonacci numbers on each call to next()
# As 'next(x)' is called, the generator continues yielding the next Fibonacci number in the sequence

# For example:
# print(next(x)) will output 5 (6th Fibonacci number)
# print(next(x)) will output 8 (7th Fibonacci number)
# print(next(x)) will output 13 (8th Fibonacci number)
# print(next(x)) will output 21 (9th Fibonacci number)
# print(next(x)) will output 34 (10th Fibonacci number)


```

```python
#6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(expo):
  """
  This function yields the powers of 2 up to a given exponent.
  """
  for i in range(1,expo+1):
    yield 2**i

x = powers_of_2(5)

print(next(x))  # 2^1 = 2
print(next(x))  # 2^2 = 4
print(next(x))  # 2^3 = 8
print(next(x))  # 2^4 = 16
print(next(x))  # 2^5 = 32




```

```python
#10 Create a Python program that uses `filter()` to remove all the vowels from a given string

def rem_vow(words):
  """
  This function removes all the vowels from a given string.
  """
  vowels = 'aeiouAEIOU'
  return ''.join(filter(lambda char : char not in vowels,words))    # Remove vowels from each word by filtering out characters that are vowels

l = ['car','Danish','Ajay sir','john']

list(map(rem_vow,l))                 #  Apply rem_vow to each word in the list using map and convert the result to a list

```
